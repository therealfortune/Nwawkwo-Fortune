```python
import random 

class TicTacToe:

    def __init__(self):
        self.board = []
        self.row = 3
        self.column = 3
        self.placeholder = '_'
        self.x = 'x'
        self.o = 'o'
        self.create_board()
        self.display_board()
        self.game()


    def __str__(self):
        return str(self.board)

    def create_board(self):
        for x in range(self.row):
            f = []
            for y in range(self.column):
                f.append(self.placeholder)
            self.board.append(f)

    def display_board(self):
        for x in self.board:
            for y in x:
                print(y, end='\t')
            print()

    def get_user_input(self):
        while True:
            n = input('Enter your coordinates e.g 3 2 : ')
            n = n.split(' ')
            if len(n) == 2:
                if n[0].isnumeric() and n[1].isnumeric():
                    if (int(n[0]) > self.row or int(n[0]) < 1) or (int(n[1]) > self.column or int(n[1]) < 1):
                        print('Enter coordinates within the range of 1 and 3')
                    else:
                        if self.board[int(n[0])-1][int(n[1])-1] == self.placeholder:
                            break
                        else:
                            print('Coordinates are already filled')
                else:
                    print('Enter numeric values')
            else:
                print('Let your coordinates be spaced')
        return n

    def add_board_input(self, coord, player=None):
        if player == 'comp':
            x,y = coord[0], coord[1]
            holder = self.o
        else:
            x, y = int(coord[0]) - 1 , int(coord[1])  - 1
            holder = self.x
        if x >= 0 and y >= 0 and holder :
            self.board[x][y] = holder

    def get_comp_input(self):
        while True:
            endpoint_x = self.row - 1
            endpoint_y = self.row - 1

            coord_x = random.randint(0, endpoint_x)
            coord_y = random.randint(1, endpoint_y)

            n = [coord_x, coord_y]
            if self.board[coord_x][coord_y] == self.placeholder:
                break
        return n

    def check_game_over(self):
        for i in self.board:
            if self.placeholder in i:
                break
        else:
            return True
        return False

    def check_winner(self):
        winner = None 
        for x in self.board:
            if x[0] == x[1] == x[2]:
                if x[0] != self.placeholder:
                    winner = x[0]
                    return winner

        x,y,z = self.board
        for a in range(self.row):
            if x[a] == y[a] == z[a]:
                if x[a] != self.placeholder:
                    winner = x[a]
                    return winner

        if x[0] == y[1] == z[2]:
            if x[0] != self.placeholder:
                winner = x[0]
                return winner

        if x[2] == y[1] == z[0]:
            if x[2] != self.placeholder:
                winner = x[2]
                return winner

    def game(self):
        game_over = False
        winner = None 

        while not game_over:
            user_input = self.get_user_input()
            self.add_board_input(user_input)
            winner = self.check_winner()
            if winner:
                self.display_board()
                if winner == self.x:
                    print('Player wuns')
                else:
                    print('Computer AI wins')
                game_over = True
                break
            
            if self.check_game_over():
                self.display_board()
                print('DRAW')
                break

            comp_input = self.get_comp_input()
            self.add_board_input(comp_input, 'comp')
            self.display_board()
            winner = self.check_winner()
            if winner:
                if winner == self.x:
                    print('You win')
                else:
                    print('Computer wins')
                game_over = True
                break
            if self.check_game_over():
                print('Draw')
                game_over = True
                break



x = TicTacToe()
                

```
